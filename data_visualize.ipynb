**Bar Charts and Heatmaps**

```python
#dataset flight delay will be used for this part
flight_delay = pd.read_csv('data/flight_delays.csv', index_col='Month')
flight_delay.head(2)
```

```python
## Bar Chart
plt.figure(figsize=(10,4))

#add title
plt.title('Avarage arrival delay for spirit Airlines')
sns.barplot(x=flight_delay.index, y=flight_delay['NK'], label='New Kingdom')
#add label for y-axis
plt.ylabel('Arrival delay (in minutes')

plt.show()
```

```python
# Heatmap plotting
#set width and height as usual
plt.figure(figsize=(14,6))
#add title
plt.title('Avarage Arrival Delay for each airline, by month')

# heatmap showing avarage dealy for each airline by month
sns.heatmap(data=flight_delay, annot=True)

#add label for x-axis
plt.xlabel('Airline')

plt.show()
```
